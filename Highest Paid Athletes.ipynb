```python

import pandas as pd
```

```python
# Wikipedia URL
url = "https://en.wikipedia.org/wiki/Forbes_list_of_the_world%27s_highest-paid_athletes"

# Extract all tables from the page
tables = pd.read_html(url)
tables
```

```python
years = list(range(2024, 2019, -1)) + list(range(2019, 2011, -1))
years
```

```python
tables_to_merge = tables[:5] + tables[6:14]
tables_to_merge
```

```python
if len(tables_to_merge) == len(years):
    for i, year in enumerate(years):
        tables_to_merge[i]['Year'] = year
    # Merge all tables into one DataFrame
    aths_df = pd.concat(tables_to_merge, join="outer", ignore_index=True)

    # Display the first few rows
    #print(merged_df.head())

else:
    print(f"Mismatch: Found {len(tables_to_merge)} tables but expected {len(years)}.")
```
